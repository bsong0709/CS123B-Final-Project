```python
data = dict()
data_meta = dict()
data_meta['714'] = m.read_meta_data('714')
data['714_Normalized'] = m.read_phenotype_data('714', 'GLDS-638_rna_seq_Normalized_Counts_GLbulkRNAseq')
```

```python
# Getting the relevent kind of data that we need
name_type = []

for i in data_meta['714']:
    if i == 'Source Name' or i == 'Sample Name':
        name_type.append(i)
    elif 'Factor Value' in i:
        name_type.append(i)

name_type

rel_data = data_meta['714'][name_type]
rel_data
```

```python
# Getting an array of gravity values in g
test_type = data_meta['714']['Factor Value[Spaceflight]']
source_name = data_meta['714']['Source Name']
grav_value = data_meta['714']['Factor Value[Altered Gravity]']

grav_list = []

for i in range(len(data_meta['714'])):
    if '1G' in grav_value[i]:
        grav_list.append(1)
    elif '1/6G' in grav_value[i]:
        grav_list.append(1/6)
    else:
        grav_list.append(0)


y = m.np.array(grav_list)

# data_meta['714'].columns.tolist()
```

```python
# Remove non-protein coding genes, drop NaN values, drop low coefficient of variance genes
df = m.filter_data(data['714_Normalized'], dropnans=True, dropgenes='non-coding', droplowcvs=0.5)
```

```python
X = list()
for sample in df.columns[1:]:
  X.append(list(df[sample]))
X = m.np.array(X)
```

```python
#Split training and testing set, 70/30 split
X_train, X_test, y_train, y_test = m.train_test_split(X, y, test_size=0.3, random_state=0)

print('X train: ', X_train.shape)
print('y train: ', len(y_train))
print('X test: ', X_test.shape)
print('y test: ', len(y_test))
```

```python
reg = m.LinearRegression().fit(X_train, y_train)
print('score: ', reg.score(X_train, y_train))
print('coefs: ', reg.coef_)
print('intercept: ', reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
pred = m.pd.DataFrame({'True_value': y_test, 'Predicted': y_pred})
pred
```

```python
m.sns.lmplot(data=pred, x='True_value', y= 'Predicted', ci=None);
```

```python
# get performance of regression
# r2 score
r2=m.r2_score(y_test, y_pred, multioutput='variance_weighted')
print('r2 score: ', r2)

# mean absolute error
mae = m.mean_absolute_error(y_test, y_pred)
print('mae: ', mae)

# mean squared error
mse = m.mean_squared_error(y_test, y_pred)
print('mse: ', mse)
```

```python
reg = m.LinearRegression().fit(X, y)
genes_coefs_dict = dict()
genes_list = list(df['Unnamed: 0'])
coefs_list = list(reg.coef_)
for i in range(len(genes_list)):
  genes_coefs_dict[genes_list[i]] = coefs_list[i]

genes_coefs_sorted_dict = {k: v for k, v in sorted(genes_coefs_dict.items(), key=lambda item: item[1], reverse=True)}

top10_genes = list(m.islice(genes_coefs_sorted_dict, 10))
print(top10_genes)
```

```python
top10_gene_symbols = m.get_symbol_from_id(top10_genes)
print(top10_gene_symbols)
```

```python
means_list = list(data['714_Normalized'].drop(columns=['Unnamed: 0']).mean(axis=1))
vars_list = list(data['714_Normalized'].drop(columns=['Unnamed: 0']).var(axis=1))

log2_means = list()
log2_vars = list()

for mean in means_list:
  log2_means.append(m.np.log2(1 + mean))

for v in vars_list:
  log2_vars.append(m.np.log2(1 + v))
```

```python
fig, ax = m.plt.subplots(figsize=(10,10))
ax.set_xlabel('mean', fontsize=18)
ax.set_ylabel('variance', fontsize=18)
m.plt.scatter(x=log2_means, y=log2_vars, marker='o', color='black', s=10)
m.plt.show()
```
